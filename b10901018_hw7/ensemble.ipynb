```python
def process_csv(path, Pth=0.5):
    p1 = []
    l1 = []
    with open(path, 'r') as f:
        for i in f:
            if(len(i.split(','))) > 3:
                str1 = ''
                for j in  i.split(',')[1:-1]:
                    str1 += j
                l1.append(str1)
                p1.append(i.split(',')[-1][:-1])
            else:
                l1.append(i.split(',')[1])
                p1.append(i.split(',')[2][:-1])
    p1 = p1[1:]
    l1 = l1[1:]

    ct = 0
    for i in range(len(p1)):
        if float(p1[i])<Pth:
            ct+=1
            l1[i] = ''
    print(f"number of lines score < {Pth} = {ct}")
    return l1, p1
```

```python
def merge(l1,p1,l2,p2):
    lm = []
    pm = []
    for i in l1:
        lm.append(i)
    for i in p1:
        pm.append(i)
    ct = 0
    cta = 0
    for i in range(len(l1)):
        if l1[i] == "" :
            cta += 1
            if l2[i] != "" :
                ct += 1
                lm[i] = l2[i]
                pm[i] = p2[i]
    print(ct, cta, ct/cta)

    return lm, pm
```

```python
def make_result(result, Fname = "result.csv"):
    print(f"number of lines = {len(result)}")
    with open(Fname, 'w') as f:
        f.write("ID,Answer\n")
        for i in range(len(result)):
            f.write(f"{i},{result[i]}\n")

    print(f"Completed! Result is in {Fname}")
```
